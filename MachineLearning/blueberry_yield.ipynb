```python
from sklearn.preprocessing import Normalizer, StandardScaler
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.compose import ColumnTransformer


norm = Normalizer()
std = StandardScaler()


pip_transf = Pipeline(steps=[('norm', norm), ('std', std)])


preprocessor = ColumnTransformer(transformers=[
    ('pipeline', pip_transf, ['fruitset', 'fruitmass', 'seeds']),
], remainder='passthrough', verbose_feature_names_out=False, n_jobs=-1)

preprocessor
```

```python
from sklearn.model_selection import train_test_split

X = train_df.drop('yield', axis=1)
y = train_df['yield']

X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.85)
print(X_train.shape)
print(X_val.shape)
```

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression, Ridge, ElasticNet, HuberRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, HistGradientBoostingRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.preprocessing import QuantileTransformer


lr_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('lr', LinearRegression())]),
    transformer=QuantileTransformer(output_distribution='normal')
    )

ridge_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('ridge', Ridge())]),
    transformer=QuantileTransformer(output_distribution='normal')
)
elastic_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('elastic', ElasticNet(max_iter=8000))]),
    transformer=QuantileTransformer(output_distribution='normal')
)
huber_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('huber', HuberRegressor(max_iter=8000))]),
    transformer=QuantileTransformer(output_distribution='normal')
)
rf_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('rf', RandomForestRegressor())]),
    transformer=QuantileTransformer(output_distribution='normal')
)
etr_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('etr', ExtraTreesRegressor())]),
    transformer=QuantileTransformer(output_distribution='normal')
)
hgbr_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('hgbr', HistGradientBoostingRegressor(max_iter=8000))]),
    transformer=QuantileTransformer(output_distribution='normal')
)
gbr_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('gbr', GradientBoostingRegressor())]),
    transformer=QuantileTransformer(output_distribution='normal')
)
svr_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('svr', SVR(max_iter=8000))]),
    transformer=QuantileTransformer(output_distribution='normal')
)
xgb_rgr = TransformedTargetRegressor(
    regressor=Pipeline(steps=[('preprocessor', preprocessor), ('xgb', XGBRegressor())]),
    transformer=QuantileTransformer(output_distribution='normal')
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score

regressors = [lr_rgr, ridge_rgr, huber_rgr, elastic_rgr, rf_rgr, etr_rgr, hgbr_rgr, gbr_rgr, svr_rgr, xgb_rgr]
model_names = ['LinearRegression', 'Ridge', 'HuberRegressor', 'ElasticNet', 'RandomForestRegressor', 'ExtraTreesRegressor', 'HistGradientBoostingRegressor', 'GradientBoostingRegressor', 'SVR', 'XGBRegressor']

results={}
for model, regressor in zip(model_names, regressors):
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_val)
    mae = np.round(mean_absolute_error(y_true=y_val, y_pred=y_pred), 3)
    mape = np.round(mean_absolute_percentage_error(y_true=y_val, y_pred=y_pred), 3)
    rmse = np.round(np.sqrt(mean_squared_error(y_true=y_val, y_pred=y_pred)), 3)
    r2 = np.round(r2_score(y_true=y_val, y_pred=y_pred))
    results[model]={'mae':mae, 'mape':mape, 'rmse':rmse, 'r2':r2}

results_df = pd.DataFrame.from_dict(results, orient='index')
results_df.sort_values('mae', ascending=True)
```

```python
import optuna
from sklearn.model_selection import cross_val_score
def objective(trial):

    hgbr_lr = trial.suggest_float('hgbr_lr', 1e-5, 1e-1)
    hgbr_max_iter = trial.suggest_int('hgbr_max_iter', 15000, 25000, step=500)
    hgbr_max_depth = trial.suggest_int('hgbr_max_depth', 20, 30, step=1)
    regressor = TransformedTargetRegressor(
        regressor=Pipeline(steps=[('preprocessor', preprocessor), ('hgbr', HistGradientBoostingRegressor(learning_rate=hgbr_lr, max_iter=hgbr_max_iter, max_depth=hgbr_max_depth))]),
        transformer=QuantileTransformer(output_distribution='normal')
    )
    

    score = cross_val_score(regressor, X_train, y_train, n_jobs=-1, cv=10, scoring='neg_mean_absolute_error')
    tot_score = -(score.mean())

    return tot_score

if __name__ == '__main__':
    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=300)
    print(study.best_trial)
```

```python
from torch import nn
from torch.utils.data import DataLoader
import torch.optim as optim
import torch
import tqdm
import copy

X_train_preprocessed = pd.DataFrame(data=preprocessor.fit_transform(X_train), 
                                    columns=preprocessor.get_feature_names_out())

X_val_preprocessed = pd.DataFrame(data=preprocessor.fit_transform(X_val), 
                                  columns=preprocessor.get_feature_names_out())

test_df_preprocessed = pd.DataFrame(data=preprocessor.fit_transform(test_df),
                                    columns=preprocessor.get_feature_names_out())
```

```python
if not torch.backends.mps.is_available():
  if not torch.backends.mps.is_built():
    print('MPS not available due to PyTorch')
  else:
    print('MPS not available due to MacOS')

else:
  print('MPS Available!')
  mps_device = torch.device('mps')
```

```python
X_train_nn = torch.tensor(X_train_preprocessed.to_numpy(), dtype=torch.float32, device='mps')
X_val_nn = torch.tensor(X_val_preprocessed.to_numpy(), dtype=torch.float32, device='mps')

y_train_nn = torch.tensor(y_train.to_numpy(), dtype=torch.float32, device='mps').reshape(-1,1)
y_val_nn = torch.tensor(y_val.to_numpy(), dtype=torch.float32, device='mps').reshape(-1,1)
```

```python
X_train_preprocessed.shape
```

```python
model = nn.Sequential(
  nn.Linear(16, 1024),
  nn.ReLU(),
  nn.Linear(1024,512),
  nn.ReLU(),
  nn.Linear(512,256),
  nn.ReLU(),
  nn.Linear(256, 128),
  nn.ReLU(),
  nn.Linear(128,64),
  nn.ReLU(),
  nn.Linear(64,32),
  nn.ReLU(),
  nn.Linear(32,1)
)

loss_fn = nn.L1Loss()
optimiser = optim.Adam(model.parameters(), lr=1e-4)

n_epochs = 15000
batch_size=12995
batch_start = torch.arange(0, len(X_train_preprocessed), batch_size)

best_error = np.inf
best_weights = None
history = []

model.to(mps_device)
for epoch in tqdm.tqdm(range(n_epochs)):
  model.train()
  with tqdm.tqdm(batch_start, unit='batch', mininterval=0, disable=True) as bar:
    bar.set_description(f'Epoch {epoch}')
    for start in bar:
      X_batch = X_train_nn[start:start+batch_size]
      y_batch = y_train_nn[start:start+batch_size]
      y_pred = model(X_batch)
      loss = loss_fn(y_pred, y_batch)
      optimiser.zero_grad()
      loss.backward()
      optimiser.step()
      bar.set_postfix(mse=float(loss))
  model.eval()
  y_pred = model(X_val_nn)
  error = loss_fn(y_pred, y_val_nn)
  error = float(error)
  history.append(error)
  if error < best_error:
    best_error = error
    best_weights = copy.deepcopy(model.state_dict())
```

```python
model.load_state_dict(best_weights)
print('MSE = %0.4f' % best_error)
print('RMSE = %0.4f' % np.sqrt(best_error))
```

```python
mpl.style.use('mocha')
plt.plot(history)
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.show()
```
